```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating Data Frame Atole 
df_atole= pd.read_csv('C:/Users/anamo/Downloads/atole.csv')
print(df_atole.head())
```

```python
#Graficando atole
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_atole, x= "Segundos", y= "Grados", hue="Grados")
plt.title("Atole Cooling")
plt.xlabel("Time (s) scale 1:15")
plt.ylabel("Degrees °C")
plt.show()
```

```python
# Creating Data Frame Water
df_agua= pd.read_csv('C:/Users/anamo/Downloads/agua.csv')
print(df_agua.head())

#Water
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_agua, x= "Segundos", y="Grados", palette= "Blues", hue= "Grados")  
plt.title("Water Cooling")
plt.xlabel("Time(s) scale 1:10")
plt.ylabel("Degrees °C")
plt.show()
```

```python
# Creating Data Frame Coffee
df_cafe= pd.read_csv('C:/Users/anamo/Downloads/cafe.csv')
print(df_cafe.head())

#Coffee
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_cafe, x= "Tiempo", y= "Grados", hue= "Grados", palette= "Oranges")
plt.title("Coffee Cooling")
plt.xlabel("Time (s) scale 1:12")
plt.ylabel("Degrees °C")
plt.show()
```
