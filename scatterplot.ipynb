```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating Data Frame Atole 
df_atole= pd.read_csv('C:/Users/anamo/Downloads/atole.csv')
df_atole.columns = ["Seconds", "Degrees"]
print(df_atole.head())
```

```python
#Atole scatterplot
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_atole, x= "Seconds", y= "Degrees", hue="Degrees")
plt.title("Atole Cooling")
plt.xlabel("Time (s) scale 1:15")
plt.ylabel("Degrees C")
plt.show()
```

```python
# Creating Data Frame Water
df_agua= pd.read_csv('C:/Users/anamo/Downloads/agua.csv')
df_agua.columns= ["Seconds",  "Degrees"]
print(df_agua.head())

#Water scatterplot
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_agua, x= "Seconds", y="Degrees", palette= "Blues", hue= "Degrees")  
plt.title("Water Cooling")
plt.xlabel("Time (s) scale 1:10")
plt.ylabel("Degrees C")
plt.show()
```

```python
# Creating Data Frame Coffee
df_cafe= pd.read_csv('C:/Users/anamo/Downloads/cafe.csv')
df_cafe.columns = ["Seconds", "Degrees"]
print(df_cafe.head())

#Coffee scatterplot 
sns.set_theme(style="whitegrid")
sns.scatterplot(data= df_cafe, x= "Seconds", y= "Degrees", hue= "Degrees", palette= "Oranges")
plt.title("Coffee Cooling")
plt.xlabel("Time (s) scale 1:12")
plt.ylabel("Degrees C")
plt.show()
```
